```python
import pandas as pd

# Load data
df = pd.read_csv("diabetes_012_health_indicators_BRFSS2015.csv")

# Convert types
binary_cols = [
    'HighBP','HighChol','CholCheck','Smoker','Stroke',
    'HeartDiseaseorAttack','PhysActivity','Fruits','Veggies',
    'HvyAlcoholConsump','AnyHealthcare','NoDocbcCost',
    'DiffWalk','Sex'
]

ordinal_cols = ['Age','Education','Income']

df[binary_cols + ordinal_cols + ['Diabetes_012']] = \
    df[binary_cols + ordinal_cols + ['Diabetes_012']].astype('int8')

# BMI outliers (safer: cap instead of drop)
df['BMI'] = df['BMI'].clip(10, 80)

# Binary target
df['Diabetes_binary'] = (df['Diabetes_012'] > 0).astype('int8')

# Quick check
print(df.shape)
print(df.dtypes.value_counts())
```
